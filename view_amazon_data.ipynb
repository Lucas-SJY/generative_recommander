```python
def load_json_or_jsonl(path):
    with open(path, "r") as f:
        txt = f.read().strip()
    # 尝试整体解析
    try:
        return json.loads(txt)
    except json.JSONDecodeError:
        # 逐行解析（跳过空行）
        return [json.loads(line) for line in txt.splitlines() if line.strip()]
```

```python
meta = load_json_or_jsonl("dataset/endomondoMeta.json")
hr = load_json_or_jsonl("dataset/endomondoHR.json")

meta_fields = set().union(*(d.keys() for d in meta))
hr_fields = set().union(*(d.keys() for d in hr))
print("Meta fields:", sorted(meta_fields))
print("HR fields:", sorted(hr_fields))
```

```python
from datasets import load_dataset

# 加载 Amazon Reviews 2023 数据集
dataset = load_dataset("McAuley-Lab/Amazon-Reviews-2023", "all_csv")

print("Dataset splits:", dataset.keys())
print("\nDataset structure:")
print(dataset)

# 显示第一个 split 的字段
first_split = list(dataset.keys())[0]
print(f"\n'{first_split}' split 字段:")
print(dataset[first_split].column_names)

# 显示数据样例
print(f"\n'{first_split}' split 前 3 行数据:")
for i in range(min(3, len(dataset[first_split]))):
    print(f"\n--- 样本 {i+1} ---")
    for key, value in dataset[first_split][i].items():
        print(f"{key}: {value}")
```
